```r
p <- ggplot(diamonds)
```

```r
p + layer(
    mapping= aes(x=carat),
    data=diamonds,
    geom = "bar",
    params = list(fill = 'darkblue'),
    stat="count",
    position='identity',
)
```

```r
p + layer(
    mapping= aes(x=carat, y=price),
    data=diamonds,
    geom = "point",
    stat="identity",
    position="identity",
) + scale_x_log10()
```

```r
p + layer(
    mapping= aes(x=carat, fill=cut),
    data=diamonds,
    geom = "bar",
    params = list(binwidth=.1),
    stat="bin",
    position='dodge',
) + scale_x_log10()
```
