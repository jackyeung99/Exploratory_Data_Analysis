### Distributions

Measuring the PDF of the feature NKP30, which looks fairly symetrical and normally distributed. There is also a small cluster on the left most cluster that does not follow the typical normal distribution. We will explore this more thouroughly with a Q-normal plot 

```r
ggplot(df, aes(x = NKp30)) + geom_histogram(bins = 30)
```

CDF of the NKP30 feature

```r
ggplot(df, aes(x = NKp30)) + stat_ecdf()
```

### QQ Plot

Here we examine the distribution of the NKP30 feature against the best fitting normal distribution. From this graph we notice small divergence form the y=x line in both the left and right most tails. 

```r
ggplot(df, aes(sample = NKp30)) + stat_qq() + stat_qq_line(color='Blue')
```

### QQ plot against another feature

The QQ plot comparing NKP30 and CXCR6 suggests that these two variables follow different distributions, as the plotted points deviate from the y=x line instead of aligning with it.









```r
ggplot(df, aes(sample = NKp30)) + stat_qq(aes(sample = CXCR6)) + stat_qq_line(color='Blue')
```
